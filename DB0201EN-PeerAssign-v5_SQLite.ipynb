```python
import csv, sqlite3
import pandas as pd
import requests

con = sqlite3.connect("FinalDB.db")
cur = con.cursor()

%sql sqlite:///FinalDB.db

df_cd = pd.read_csv('ChicagoCensusData.csv')
df_cd.to_sql("CENSUS_DATA", con, if_exists='replace', index=False,method="multi")

df_cps = pd.read_csv('ChicagoPublicSchools.csv')
df_cps.to_sql("CHICAGO_PUBLIC_SCHOOLS", con, if_exists='replace', index=False,method="multi")

df_ccd = pd.read_csv('ChicagoCrimeData.csv')
df_ccd.to_sql("CHICAGO_CRIME_DATA", con, if_exists='replace', index=False,method="multi")
```

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
cur.execute('select count(*) from CHICAGO_CRIME_DATA')
print("Nb Crimes : ", cur.fetchall()[0][0])
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
cur.execute('select community_area_name from CENSUS_DATA where per_capita_income < 11000 ')
pd.DataFrame(cur.fetchall(),columns =['community_area_name'])
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
df_ccd.columns
```

```python
cur.execute('select ccd.CASE_NUMBER from CHICAGO_CRIME_DATA as ccd, CENSUS_DATA as cd on cd.community_area_number = ccd.community_area_number where cd.PERCENT_AGED_UNDER_18_OR_OVER_64 >0')
pd.DataFrame(cur.fetchall(),columns=['CASE_NUMBER'])
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
cur.execute('select ccd.* from CHICAGO_CRIME_DATA as ccd, CENSUS_DATA as cd on cd.community_area_number = ccd.community_area_number where cd.PERCENT_AGED_UNDER_18_OR_OVER_64 >0 and ccd.PRIMARY_TYPE like "KID%"')
column_names = [description[0] for description in cur.description]
pd.DataFrame(cur.fetchall(),columns=column_names)
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
cur.execute('select distinct PRIMARY_TYPE from CHICAGO_CRIME_DATA where LOCATION_DESCRIPTION like "SCHOOL%"')
pd.DataFrame(cur.fetchall(),columns=['PRIMARY_TYPE'])
```

### Problem 6

##### List the average safety score for each type of school.


```python
cur.execute('select "Elementary, Middle, or High School", avg(SAFETY_SCORE) from CHICAGO_PUBLIC_SCHOOLS group by("Elementary, Middle, or High School")')
pd.DataFrame(cur.fetchall(), columns=['School_Type', 'Avg_Safety_Score'])
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
cur.execute('select COMMUNITY_AREA_NAME from CENSUS_DATA order by(PERCENT_HOUSEHOLDS_BELOW_POVERTY) desc limit 5')
pd.DataFrame(cur.fetchall(), columns=['COMMUNITY_AREA_NAME'])
```

### Problem 8

##### Which community area is most crime prone?


```python
cur.execute(' select COMMUNITY_AREA_NAME from CENSUS_DATA order by PERCENT_OF_HOUSING_CROWDED desc limit 1')
pd.DataFrame(cur.fetchall(), columns=['COMMUNITY_AREA_NAME'])
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
cur.execute('select COMMUNITY_AREA_NAME from CENSUS_DATA where HARDSHIP_INDEX = (select max(HARDSHIP_INDEX) from CENSUS_DATA)')
pd.DataFrame(cur.fetchall(), columns=['COMMUNITY_AREA_NAME'])
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
cur.execute('select cd.COMMUNITY_AREA_NAME from CENSUS_DATA as cd JOIN (select ccd.COMMUNITY_AREA_NUMBER, count(*) as crime_count from CHICAGO_CRIME_DATA as ccd group by ccd.COMMUNITY_AREA_NUMBER order by crime_count desc limit  1) as max on cd.COMMUNITY_AREA_NUMBER = max.COMMUNITY_AREA_NUMBER')
pd.DataFrame(cur.fetchall(), columns=['COMMUNITY_AREA_NAME'])
```
